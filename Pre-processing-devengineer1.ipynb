```python
import re
from bs4 import BeautifulSoup

def extract_text(text, del_tag = ['a']):                                                          #1
    if type(text) == str:
        soup = BeautifulSoup(text, 'lxml')
        for elem in [soup.find_all(tag) for tag in del_tag]:                                      #2
            for sub_elem in elem:
                sub_elem.decompose()
        text = ' '.join([tag.text for tag in soup.find_all('body')])                              #3
        text = re.sub('\s+', ' ', text)
        return text.strip().lower()
    else:
        return ' '
```

```python
data["combined_new"] = data['combined_new'].apply(extract_text, args=(['a', 'table'], ))
```

```python
import re

def junk_in(text, starts_with, ends_with):
    idxs1 = [text.find(word) for word in starts_with]                                                   #1
    while sum(idxs1) != -1*len(starts_with):                                                            #2
        start = [i for i in sorted(idxs1) if i >= 0][0]                                                 #3
        idxs2 = [text.find(word, start + 1) for word in ends_with]                                      #4
        if sum(idxs2) != -1*len(ends_with):                                                             #5
            end = [i for i in sorted(idxs2) if i >= 0][0]                                               #6
        else:                                                                                           #7
            end = None                                                                                  #8
        rem = text[start:end]                                                                           #9
        print(rem)                                                                                      #10
        text = text.replace(rem, ' ')                                                                   #11
        idxs1 = [text.find(word) for word in starts_with]                                               #12
    print(re.findall('<.*?>|\{.*?\}', text))                                                            #13
    print(re.findall('(\S+@\S+\.\S+|\w\d{4}"\ssrc=.*?>|http\S+)', text))                                #14
    print(re.findall('_+|-+|\*+', text))                                                                #15
    print('------------------------------------------------------------------------------------------------------------------')
def junk_out(text, starts_with, ends_with):
    idxs1 = [text.find(word) for word in starts_with]
    
    while sum(idxs1) != -1*len(starts_with):
        start = [i for i in sorted(idxs1) if i >= 0][0]
        idxs2 = [text.find(word, start) for word in ends_with]
        if sum(idxs2) != -1*len(ends_with):
            end = [i for i in sorted(idxs2) if i >= 0][0]
        else:
            end = None
        rem = text[start:end]
        text = text.replace(rem, ' ')
        idxs1 = [text.find(word) for word in starts_with]
    text = re.sub('<.*?>|\{.*?\}', ' ', text)
    text = re.sub('(\S+@\S+\.\S+|\w\d{4}"\ssrc=.*?>|http\S+)', ' ', text)
    text = re.sub('_+|-+|\*+', ' ', text)
    text = re.sub('\s+', ' ', text).strip()
    return text    
```

```python
#removing stopwords
import nltk
from nltk.tokenize import word_tokenize

def rem_stopwords(strng, custom_stopwords, min_len=None):
    if type(strng) == str:
        word_tokens = word_tokenize(strng)
        if min_len != None:
            return ' '.join([w for w in word_tokens if w not in custom_stopwords and len(w) > min_len]).strip()
        else:
            return ' '.join([w for w in word_tokens if w not in custom_stopwords]).strip()
    else:
        return ' '
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
stops = stopwords.words('english') + ends_with + additional
```

```python
data["combined_new"] = data["combined_new"].apply(rem_stopwords, args=(stops, ))
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

def rem_nums(text, lemmatization= True):
    text = re.sub('[a-z]{2}\d+\s', ' ', text)
    text = re.sub('[^a-zA-Z]', ' ', text)
    text = re.sub(' +', ' ', text).strip()
    if lemmatization == True:
        return ' '.join(WordNetLemmatizer().lemmatize(w) for w in word_tokenize(text) if len(w) > 1)
    else:
        return ' '.join(w for w in word_tokenize(text) if len(w) > 1)
```

```python
data["combined_new"] = data["combined_new"].map(rem_nums)
```
